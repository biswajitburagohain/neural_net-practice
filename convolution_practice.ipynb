```
from matplotlib import pyplot as plt
from matplotlib import image
import numpy as np
```

```
img_dir = "https://www.science-emergence.com/media/images/src/459.png"
img = image.imread(img_dir)
img = img[:,:,0]
print(img.shape)
plt.imshow(img)
plt.show()
```

```
trans = np.copy(img)
img = np.asarray(img)
```

```
x_axis = trans.shape[0]
y_axis = trans.shape[1]
```

```
#filter = [[0, 1, 0], [1, -4, 1], [0, 1, 0]]
#filter = [ [-1, -2, -1], [0, 0, 0], [1, 2, 1]]
filter = [ [-1, 0, 1], [-2, 0, 2], [-1, 0, 1]]
weight = 1
#convolutions
```

```
for x in range(1, x_axis-1):
  for y in range(1, y_axis-1):
    convolutions = 0.0
    convolutions = convolutions + (img[x-1,y-1] * filter[0][0])
    convolutions = convolutions + (img[x, y-1] * filter[0][1])
    convolutions = convolutions + (img[x+1, y-1] * filter[0][2])
    convolutions = convolutions + (img[x-1, y] * filter[1][0])
    convolutions = convolutions + (img[x, y] * filter[1][1])
    convolutions = convolutions + (img[x+1, y] * filter[1][2])
    convolutions = convolutions + (img[x-1, y+1] * filter[2][0])
    convolutions = convolutions + (img[x, y+1] * filter[2][1])
    convolutions = convolutions + (img[x+1, y+1] * filter[2][2])
    convolutions = convolutions * weight
    if(convolutions < 0):
      convolutions = 0
    if(convolutions>255):
      convolutions = 255
    trans[x+1,y+1] = convolutions
```

```
plt.imshow(trans)
plt.show
```

```
new_x = int(x_axis/2)
new_y = int(y_axis/2)
newimage = np.zeros((new_x, new_y))
for x in range(0, x_axis, 2):
  for y in range(0, y_axis,2):
    pixel= []
    pixel.append(trans[x,y])
    pixel.append(trans[x,y+1])
    pixel.append(trans[x+1, y])
    pixel.append(trans[x+1, y+1])
    newimage[int(x/2),int(y/2)] = max(pixel)

plt.imshow(newimage)
plt.show()
```
